# Plot the number of meetings for filered Think Tanks 

```python
import pandas as pd
df = pd.read_pickle('/mnt/2To/jupyter_data/FTM/ThinkTank/NEW_EU_TT.pkl')
```

```python
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
import matplotlib.pyplot as plt

# New graph for Commission Meetings
fig_comm, ax_comm = plt.subplots(figsize=(7, 5))
ax_comm.hist(df['nb_comm_meet'].astype(float), bins=20, alpha=0.5, log=True)
ax_comm.set_title('Commission Meetings Histogram')
ax_comm.set_xlabel('Number of Commission Meetings')
ax_comm.set_ylabel('Number of Think Tanks (log)')
axins_comm = zoomed_inset_axes(ax_comm, 2, loc='upper right')
axins_comm.hist(df['nb_comm_meet'].astype(float), bins=20, alpha=0.5)
axins_comm.set_xlim(0, 20)
axins_comm.set_ylim(8, 100)
plt.xticks(visible=True)
plt.yticks(visible=True)
mark_inset(ax_comm, axins_comm, loc1=2, loc2=4, fc="none", ec="0.5")

# New graph for Parliament Meetings
fig_parl, ax_parl = plt.subplots(figsize=(7, 5))
ax_parl.hist(df['nb_parl_meet'].astype(float), bins=20, alpha=0.5, log=True)
ax_parl.set_title('Parliament Meetings Histogram')
ax_parl.set_xlabel('Number of Parliament Meetings')
ax_parl.set_ylabel('Number of Think Tanks (log)')
axins_parl = zoomed_inset_axes(ax_parl, 2, loc='upper right')
axins_parl.hist(df['nb_parl_meet'].astype(float), bins=20, alpha=0.5, log=False)
axins_parl.set_xlim(0, 5)
axins_parl.set_ylim(8, 100)
plt.xticks(visible=True)
plt.yticks(visible=True)
mark_inset(ax_parl, axins_parl, loc1=2, loc2=4, fc="none", ec="0.5")

plt.show()
plt.show()
# END: Add zoomed-in plot
```
